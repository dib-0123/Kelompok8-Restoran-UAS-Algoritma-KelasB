```
Pelanggan Datang & Memesan
        |
        v
   [ QUEUE ]  Antrean pesanan masuk (FIFO) - menunggu diproses staff
        |
        v
   [ BST ]    Pesanan disimpan sebagai data aktif (key = ID Pesanan)
```

#### 1. Struktur Data: QUEUE (Antrean Pesanan Masuk)

Diimplementasikan menggunakan **Linked List** agar operasi `enqueue` dan `dequeue` keduanya O(1) (tidak perlu menggeser elemen seperti pada array/list biasa).

**Digunakan untuk:** menampung pesanan yang baru masuk dari pelanggan sebelum diproses oleh staff, sesuai urutan kedatangan (FIFO).

````python
# ==========================================================
# STRUKTUR DATA 1: QUEUE (Antrean Pesanan Masuk - FIFO)
# Diimplementasikan dari nol menggunakan Linked List
# (TIDAK menggunakan queue / collections.deque)
# ==========================================================

class _NodeQueue:
    """Node internal untuk linked list Queue."""
    def __init__(self, data):
        self.data = data
        self.next = None


class Queue:
    """
    Queue FIFO (First In First Out) berbasis Linked List.
    Digunakan untuk mengelola pesanan yang baru masuk dari pelanggan,
    menunggu untuk diproses oleh dapur/kasir sesuai urutan kedatangan.

    Kompleksitas:
        enqueue : O(1)
        dequeue : O(1)
        peek    : O(1)
        display : O(n)
    """

    def __init__(self):
        self._head = None   # depan antrean (yang paling lama)
        self._tail = None   # belakang antrean (yang paling baru)
        self._size = 0

    def is_empty(self):
        return self._size == 0

    def size(self):
        return self._size

    def enqueue(self, data):
        """Menambahkan data baru ke belakang antrean. O(1)"""
        node = _NodeQueue(data)
        if self.is_empty():
            self._head = node
            self._tail = node
        else:
            self._tail.next = node
            self._tail = node
        self._size += 1

    def dequeue(self):
        """Mengambil & menghapus data dari depan antrean. O(1)"""
        if self.is_empty():
            return None
        node = self._head
        self._head = self._head.next
        if self._head is None:
            self._tail = None
        self._size -= 1
        return node.data

    def peek(self):
        """Melihat data paling depan tanpa menghapus. O(1)"""
        if self.is_empty():
            return None
        return self._head.data

    def display(self):
        """Menampilkan seluruh isi antrean dari depan ke belakang. O(n)"""
        if self.is_empty():
            print("   (Antrean kosong)")
            return
        current = self._head
        idx = 1
        while current is not None:
            print(f"   {idx}. {current.data}")
            current = current.next
            idx += 1
````

#### 2. Struktur Data: STACK (Riwayat Transaksi)

Diimplementasikan menggunakan **Linked List** (penyisipan di head) sehingga `push` dan `pop` O(1).

**Digunakan untuk:** mencatat riwayat transaksi (pesanan Selesai / Dibatalkan) sehingga transaksi paling akhir dapat dilihat atau dibatalkan (undo) lebih dulu (LIFO).

````python
# ==========================================================
# STRUKTUR DATA 2: STACK (Riwayat Transaksi / Undo - LIFO)
# Diimplementasikan dari nol menggunakan Linked List
# ==========================================================

class _NodeStack:
    """Node internal untuk linked list Stack."""
    def __init__(self, data):
        self.data = data
        self.next = None


class Stack:
    """
    Stack LIFO (Last In First Out) berbasis Linked List.
    Digunakan untuk mencatat riwayat transaksi (pesanan selesai / dibatalkan)
    sehingga transaksi paling akhir bisa dilihat atau di-undo lebih dulu.

    Kompleksitas:
        push    : O(1)
        pop     : O(1)
        peek    : O(1)
        display : O(n)
    """

    def __init__(self):
        self._top = None
        self._size = 0

    def is_empty(self):
        return self._size == 0

    def size(self):
        return self._size

    def push(self, data):
        """Menambahkan data baru ke puncak stack. O(1)"""
        node = _NodeStack(data)
        node.next = self._top
        self._top = node
        self._size += 1

    def pop(self):
        """Mengambil & menghapus data dari puncak stack. O(1)"""
        if self.is_empty():
            return None
        node = self._top
        self._top = self._top.next
        self._size -= 1
        return node.data

    def peek(self):
        """Melihat data paling atas tanpa menghapus. O(1)"""
        if self.is_empty():
            return None
        return self._top.data

    def display(self):
        """Menampilkan seluruh isi stack dari paling atas (terbaru) ke bawah. O(n)"""
        if self.is_empty():
            print("   (Riwayat kosong)")
            return
        current = self._top
        idx = 1
        while current is not None:
            print(f"   {idx}. {current.data}")
            current = current.next
            idx += 1
````

#### 3. Struktur Data: BINARY SEARCH TREE (Pesanan Aktif)

Di-key oleh **ID Pesanan**. Digunakan sebagai basis data utama pesanan yang sedang aktif (sudah diproses staff, belum selesai/dibatalkan), sehingga pencarian, penambahan, dan penghapusan pesanan berdasarkan ID dapat dilakukan efisien (rata-rata O(log n)).

**Traversal yang dipilih: INORDER.** Karena BST di-key oleh ID Pesanan, traversal inorder akan selalu menghasilkan daftar pesanan terurut menaik berdasarkan ID — cocok untuk menampilkan daftar pesanan aktif secara rapi dan konsisten kepada staff. Urutan kedatangan (FIFO) sudah ditangani Queue, dan urutan prioritas dapur sudah ditangani Heap, sehingga BST tidak perlu mengurus urutan itu — cukup mengurus pencarian & manajemen data berdasarkan ID.

````python
# ==========================================================
# STRUKTUR DATA 3: BINARY SEARCH TREE (Penyimpanan Pesanan Aktif)
# Diimplementasikan dari nol menggunakan Node beranotasi key-value
# ==========================================================

class _NodeBST:
    """Node internal BST. key = ID Pesanan, value = objek Pesanan."""
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.left = None
        self.right = None


class BST:
    """
    Binary Search Tree, di-key oleh ID Pesanan.
    Digunakan sebagai basis data utama pesanan yang sedang aktif
    (sudah masuk dapur, belum selesai/dibatalkan), sehingga pencarian,
    penambahan, dan penghapusan pesanan bisa dilakukan efisien
    berdasarkan ID pesanan.

    Traversal yang dipakai: INORDER.
    Alasan: inorder pada BST yang di-key oleh ID pesanan akan selalu
    menghasilkan urutan ID menaik (ascending) -- cocok untuk kebutuhan
    bisnis restoran menampilkan daftar pesanan aktif secara berurutan
    dan rapi (bukan berdasarkan urutan penyisipan atau prioritas dapur,
    yang mana sudah ditangani terpisah oleh Queue dan Heap).

    Kompleksitas (n = jumlah node, h = tinggi tree):
        insert  : O(h)  -> rata-rata O(log n), terburuk O(n) jika tree miring
        search  : O(h)
        delete  : O(h)
        inorder : O(n)
        height  : O(n)
        count   : O(n)
    """

    def __init__(self):
        self._root = None
        self._count = 0

    def is_empty(self):
        return self._root is None

    # ---------------- INSERT ----------------
    def insert(self, key, value):
        self._root = self._insert_rec(self._root, key, value)

    def _insert_rec(self, node, key, value):
        if node is None:
            self._count += 1
            return _NodeBST(key, value)
        if key < node.key:
            node.left = self._insert_rec(node.left, key, value)
        elif key > node.key:
            node.right = self._insert_rec(node.right, key, value)
        else:
            # ID sudah ada -> update datanya saja
            node.value = value
        return node

    # ---------------- SEARCH ----------------
    def search(self, key):
        return self._search_rec(self._root, key)

    def _search_rec(self, node, key):
        if node is None:
            return None
        if key == node.key:
            return node.value
        if key < node.key:
            return self._search_rec(node.left, key)
        return self._search_rec(node.right, key)

    # ---------------- DELETE ----------------
    def delete(self, key):
        self._root, deleted = self._delete_rec(self._root, key)
        return deleted

    def _delete_rec(self, node, key):
        if node is None:
            return node, False

        deleted = False
        if key < node.key:
            node.left, deleted = self._delete_rec(node.left, key)
        elif key > node.key:
            node.right, deleted = self._delete_rec(node.right, key)
        else:
            deleted = True
            # Kasus 1: tidak ada anak / hanya 1 anak
            if node.left is None:
                self._count -= 1
                return node.right, deleted
            elif node.right is None:
                self._count -= 1
                return node.left, deleted
            # Kasus 2: punya 2 anak -> cari successor (terkecil di subtree kanan)
            successor = node.right
            while successor.left is not None:
                successor = successor.left
            node.key, node.value = successor.key, successor.value
            node.right, _ = self._delete_rec(node.right, successor.key)
            self._count -= 1  # node asli tetap dihitung terhapus (digantikan successor)

        return node, deleted

    # ---------------- TRAVERSAL: INORDER ----------------
    def inorder(self):
        """Mengembalikan list value hasil traversal inorder (ID menaik)."""
        result = []
        self._inorder_rec(self._root, result)
        return result

    def _inorder_rec(self, node, result):
        if node is not None:
            self._inorder_rec(node.left, result)
            result.append(node.value)
            self._inorder_rec(node.right, result)

    # ---------------- HEIGHT & COUNT ----------------
    def height(self):
        return self._height_rec(self._root)

    def _height_rec(self, node):
        if node is None:
            return -1  # tinggi tree kosong = -1, node tunggal = 0
        return 1 + max(self._height_rec(node.left), self._height_rec(node.right))

    def count_nodes(self):
        return self._count
````

#### 4. Struktur Data: BINARY HEAP (Prioritas Dapur)

Diimplementasikan sebagai **Min-Heap berbasis array** (list Python biasa sebagai representasi array, bukan modul `heapq`). Elemen heap berupa tuple `(prioritas, urutan_masuk, pesanan)`.

**Digunakan untuk:** menentukan urutan pesanan mana yang dimasak dapur lebih dulu. Pesanan **Take Away Express** diberi prioritas 1 (lebih tinggi/didahulukan) dibanding **Dine In** yang diberi prioritas 2, merepresentasikan kebutuhan bisnis nyata restoran yang biasanya memprioritaskan pesanan take-away agar pelanggan tidak menunggu lama. Jika prioritas sama, pesanan yang lebih dulu masuk (`urutan_masuk` lebih kecil) tetap didahulukan agar adil (FIFO di antara prioritas yang sama).

````python
# ==========================================================
# STRUKTUR DATA 4: BINARY HEAP (Prioritas Proses Dapur)
# Diimplementasikan dari nol menggunakan array (list Python biasa)
# TIDAK menggunakan modul heapq / PriorityQueue
# ==========================================================

class BinaryHeap:
    """
    Min-Heap berbasis array, di-key oleh tuple (prioritas, urutan_masuk).
    Semakin KECIL nilai prioritas, semakin dulu dimasak oleh dapur.

    Digunakan untuk menentukan urutan proses masak di dapur:
        - Take Away Express -> prioritas = 1 (didahulukan)
        - Dine In            -> prioritas = 2
    Jika prioritas sama, pesanan yang lebih dulu masuk (urutan_masuk
    lebih kecil) didahulukan -> menjaga keadilan FIFO di antara
    pesanan berprioritas sama.

    Elemen heap berupa tuple: (prioritas, urutan_masuk, objek_pesanan)

    Kompleksitas (n = jumlah elemen di heap):
        insert       : O(log n)
        delete_root  : O(log n)
        peek         : O(1)
        heapify_up   : O(log n)
        heapify_down : O(log n)
        display      : O(n)
    """

    def __init__(self):
        self._data = []  # array biasa sebagai representasi heap

    def is_empty(self):
        return len(self._data) == 0

    def size(self):
        return len(self._data)

    @staticmethod
    def _parent(i):
        return (i - 1) // 2

    @staticmethod
    def _left(i):
        return 2 * i + 1

    @staticmethod
    def _right(i):
        return 2 * i + 2

    def _swap(self, i, j):
        self._data[i], self._data[j] = self._data[j], self._data[i]

    # ---------------- INSERT ----------------
    def insert(self, item):
        """item = (prioritas, urutan_masuk, objek_pesanan). O(log n)"""
        self._data.append(item)
        self._heapify_up(len(self._data) - 1)

    def _heapify_up(self, i):
        while i > 0 and self._data[i][:2] < self._data[self._parent(i)][:2]:
            self._swap(i, self._parent(i))
            i = self._parent(i)

    # ---------------- DELETE ROOT ----------------
    def delete_root(self):
        """Menghapus & mengembalikan elemen prioritas tertinggi (akar). O(log n)"""
        if self.is_empty():
            return None
        root = self._data[0]
        last = self._data.pop()
        if self._data:
            self._data[0] = last
            self._heapify_down(0)
        return root

    def _heapify_down(self, i):
        n = len(self._data)
        while True:
            l, r = self._left(i), self._right(i)
            smallest = i
            if l < n and self._data[l][:2] < self._data[smallest][:2]:
                smallest = l
            if r < n and self._data[r][:2] < self._data[smallest][:2]:
                smallest = r
            if smallest == i:
                break
            self._swap(i, smallest)
            i = smallest

    # ---------------- PEEK ----------------
    def peek(self):
        """Melihat elemen prioritas tertinggi tanpa menghapus. O(1)"""
        if self.is_empty():
            return None
        return self._data[0]

    # ---------------- REMOVE BY ID (untuk kasus pembatalan pesanan) ----------------
    def remove_by_id(self, pesanan_id):
        """
        Menghapus pesanan tertentu dari tengah heap berdasarkan ID
        (dibutuhkan saat pesanan dibatalkan sebelum sempat dimasak).
        O(n) karena perlu mencari posisinya dulu (bukan operasi heap standar).
        """
        for i, (_, _, pesanan) in enumerate(self._data):
            if pesanan.id == pesanan_id:
                last = self._data.pop()
                if i < len(self._data):
                    self._data[i] = last
                    self._heapify_down(i)
                    self._heapify_up(i)
                return True
        return False

    # ---------------- DISPLAY ----------------
    def display(self):
        """Menampilkan seluruh isi heap (urutan array, bukan urutan proses). O(n)"""
        if self.is_empty():
            print("   (Antrean dapur kosong)")
            return
        for idx, (prioritas, _, pesanan) in enumerate(self._data):
            label = "Take Away Express" if prioritas == 1 else "Dine In"
            print(f"   [{idx}] ID#{pesanan.id} - {pesanan.nama_pelanggan} "
                  f"({label}, prioritas={prioritas})")
````

#### 5. Kelas `Pesanan` (Kontrak Data)

Merepresentasikan satu objek pesanan yang mengalir melalui Queue -> BST -> Heap -> Stack.

````python
# ==========================================================
# KELAS PESANAN (Kontrak Data)
# ==========================================================

class Pesanan:
    """Merepresentasikan satu pesanan pelanggan di restoran."""

    def __init__(self, id, nama_pelanggan, jenis, menu, meja=None):
        self.id = id
        self.nama_pelanggan = nama_pelanggan
        self.jenis = jenis                # "Dine In" / "Take Away Express"
        self.menu = menu                  # list[str]
        self.meja = meja                  # nomor meja (None jika Take Away)
        self.prioritas = 1 if jenis == "Take Away Express" else 2
        self.status = "Menunggu Antre"    # Menunggu Antre / Diproses Dapur / Selesai / Dibatalkan

    def ringkas(self):
        meja_info = f", Meja {self.meja}" if self.meja else ""
        menu_str = ", ".join(self.menu)
        return (f"ID#{self.id} | {self.nama_pelanggan} ({self.jenis}{meja_info}) "
                f"| Menu: {menu_str} | Status: {self.status}")

    def __repr__(self):
        return self.ringkas()
````

#### 6. Integrasi Sistem: `SistemRestoran`

Kelas ini menyatukan keempat struktur data ke dalam satu alur proses bisnis restoran yang saling terhubung, sesuai diagram alur di atas.

````python
# ==========================================================
# INTEGRASI SISTEM: SistemRestoran
# Menggabungkan Queue, BST, Binary Heap, dan Stack dalam satu
# alur proses bisnis restoran.
#
# ALUR BISNIS:
#   Pelanggan Order
#        |
#        v
#   [QUEUE]  Antrean pesanan masuk (FIFO, tunggu diproses staff)
#        |  (staff proses antrean satu per satu)
#        v
#   [BST]    Disimpan sbg pesanan aktif (key = ID) utk pencarian cepat
#        |  (bersamaan, pesanan didaftarkan ke heap prioritas dapur)
#        v
#   [HEAP]   Antrean prioritas dapur (Take Away Express didahulukan)
#        |  (dapur ambil prioritas tertinggi utk dimasak)
#        v
#   [STACK]  Riwayat transaksi (Selesai / Dibatalkan) -> mendukung undo
# ==========================================================

class SistemRestoran:
    def __init__(self, nama_restoran="Warung Algoritma"):
        self.nama_restoran = nama_restoran
        self.antrean_masuk = Queue()        # Queue: pesanan baru menunggu diproses
        self.pesanan_aktif = BST()          # BST: pesanan yang sudah diproses & aktif
        self.antrean_dapur = BinaryHeap()   # Heap: prioritas masak dapur
        self.riwayat = Stack()              # Stack: riwayat selesai/dibatalkan (utk undo)

        self._next_id = 1
        self._urutan_masuk = 0  # counter tie-breaker untuk heap

    # ---------- 1. Pelanggan datang & pesan ----------
    def tambah_pesanan(self, nama_pelanggan, jenis, menu, meja=None):
        pesanan = Pesanan(self._next_id, nama_pelanggan, jenis, menu, meja)
        self._next_id += 1
        self.antrean_masuk.enqueue(pesanan)
        print(f"[QUEUE] Pesanan {pesanan.id} milik {nama_pelanggan} masuk antrean.")
        return pesanan

    # ---------- 2. Proses antrean masuk -> BST + Heap ----------
    def proses_antrean(self):
        pesanan = self.antrean_masuk.dequeue()
        if pesanan is None:
            print("[QUEUE] Antrean pesanan masuk kosong, tidak ada yang diproses.")
            return None

        pesanan.status = "Diproses Dapur"
        self.pesanan_aktif.insert(pesanan.id, pesanan)

        self._urutan_masuk += 1
        self.antrean_dapur.insert((pesanan.prioritas, self._urutan_masuk, pesanan))

        print(f"[BST]  Pesanan {pesanan.id} disimpan sbg pesanan aktif.")
        print(f"[HEAP] Pesanan {pesanan.id} masuk antrean prioritas dapur "
              f"({pesanan.jenis}).")
        return pesanan

    # ---------- 3. Lihat pesanan aktif (BST inorder) ----------
    def lihat_pesanan_aktif(self):
        hasil = self.pesanan_aktif.inorder()
        if not hasil:
            print("   (Belum ada pesanan aktif)")
            return
        print("   Daftar pesanan aktif (urut ID / inorder BST):")
        for p in hasil:
            print(f"   - {p.ringkas()}")

    # ---------- 4. Cari pesanan by ID (BST search) ----------
    def cari_pesanan(self, id_pesanan):
        pesanan = self.pesanan_aktif.search(id_pesanan)
        if pesanan is None:
            print(f"   Pesanan ID#{id_pesanan} tidak ditemukan di pesanan aktif.")
        else:
            print(f"   Ditemukan -> {pesanan.ringkas()}")
        return pesanan

    # ---------- 5. Dapur proses prioritas tertinggi (Heap delete_root) ----------
    def proses_prioritas_dapur(self):
        item = self.antrean_dapur.delete_root()
        if item is None:
            print("[HEAP] Antrean dapur kosong, tidak ada yang diproses.")
            return None

        _, _, pesanan = item
        pesanan.status = "Selesai"
        self.pesanan_aktif.delete(pesanan.id)
        self.riwayat.push(("Selesai", pesanan))

        print(f"[HEAP]  Dapur memproses pesanan {pesanan.id} ({pesanan.jenis}).")
        print(f"[BST]   Pesanan {pesanan.id} dihapus dari daftar aktif (selesai).")
        print(f"[STACK] Pesanan {pesanan.id} dicatat ke riwayat sbg Selesai.")
        return pesanan

    # ---------- 6. Batalkan pesanan (BST delete + Heap remove + Stack push) ----------
    def batalkan_pesanan(self, id_pesanan):
        pesanan = self.pesanan_aktif.search(id_pesanan)
        if pesanan is None:
            print(f"   Pesanan ID#{id_pesanan} tidak ditemukan / sudah tidak aktif.")
            return None

        self.pesanan_aktif.delete(id_pesanan)
        self.antrean_dapur.remove_by_id(id_pesanan)
        pesanan.status = "Dibatalkan"
        self.riwayat.push(("Dibatalkan", pesanan))

        print(f"[BST]   Pesanan {id_pesanan} dihapus dari daftar aktif.")
        print(f"[HEAP]  Pesanan {id_pesanan} dihapus dari antrean dapur (jika ada).")
        print(f"[STACK] Pembatalan pesanan {id_pesanan} dicatat ke riwayat.")
        return pesanan

    # ---------- 7. Undo pembatalan terakhir (Stack peek + pop) ----------
    def undo_pembatalan_terakhir(self):
        top = self.riwayat.peek()
        if top is None:
            print("[STACK] Riwayat kosong, tidak ada yang bisa di-undo.")
            return None

        aksi, pesanan = top
        if aksi != "Dibatalkan":
            print(f"[STACK] Item riwayat teratas adalah '{aksi}', bukan pembatalan. "
                  f"Undo hanya berlaku untuk pembatalan terakhir.")
            return None

        self.riwayat.pop()
        pesanan.status = "Diproses Dapur"
        self.pesanan_aktif.insert(pesanan.id, pesanan)
        self._urutan_masuk += 1
        self.antrean_dapur.insert((pesanan.prioritas, self._urutan_masuk, pesanan))

        print(f"[STACK] Undo pembatalan pesanan {pesanan.id}.")
        print(f"[BST]   Pesanan {pesanan.id} dikembalikan ke daftar aktif.")
        print(f"[HEAP]  Pesanan {pesanan.id} dikembalikan ke antrean dapur.")
        return pesanan

    # ---------- 8. Lihat riwayat transaksi (Stack display) ----------
    def lihat_riwayat(self):
        if self.riwayat.is_empty():
            print("   (Riwayat kosong)")
            return
        print("   Riwayat transaksi (dari yang terbaru):")
        current = self.riwayat._top
        idx = 1
        while current is not None:
            aksi, pesanan = current.data
            print(f"   {idx}. [{aksi}] {pesanan.ringkas()}")
            current = current.next
            idx += 1

    # ---------- 9. Statistik BST & tampilan Heap ----------
    def lihat_statistik(self):
        print(f"   Jumlah pesanan aktif (BST) : {self.pesanan_aktif.count_nodes()}")
        print(f"   Tinggi BST                 : {self.pesanan_aktif.height()}")
        print(f"   Jumlah pesanan di antrean masuk (Queue) : {self.antrean_masuk.size()}")
        print(f"   Jumlah pesanan di antrean dapur (Heap)  : {self.antrean_dapur.size()}")
        print(f"   Jumlah riwayat transaksi (Stack)        : {self.riwayat.size()}")
        print()
        print("   Isi antrean dapur (Heap) saat ini:")
        self.antrean_dapur.display()
````

#### 8. Demo / Pengujian Otomatis (Opsional)

Cell berikut menjalankan skenario pengujian otomatis (tanpa perlu input manual) untuk mendemonstrasikan seluruh alur bisnis sekaligus, termasuk skenario tepi (edge case) seperti: antrean kosong, ID tidak ditemukan, dan pembatalan yang di-undo.

*(Jalankan cell ini secara terpisah / independen dari menu interaktif di atas.)*

````python
sistem = SistemRestoran("Warung Algoritma")

def section(title):
    print("\n" + "#" * 60)
    print("#", title)
    print("#" * 60)

# ---- 1. Tambah beberapa pesanan ----
section("TAMBAH PESANAN (QUEUE)")
sistem.tambah_pesanan("Andi", "Dine In", ["Nasi Goreng", "Es Teh"], meja="A1")
sistem.tambah_pesanan("Budi", "Take Away Express", ["Ayam Geprek"])
sistem.tambah_pesanan("Citra", "Dine In", ["Mie Ayam"], meja="B2")
sistem.tambah_pesanan("Dedi", "Take Away Express", ["Bakso"])
print("Ukuran antrean masuk:", sistem.antrean_masuk.size())

# ---- 2. Proses semua antrean masuk -> BST + Heap ----
section("PROSES ANTREAN -> BST & HEAP")
for _ in range(4):
    sistem.proses_antrean()

# ---- 3. Lihat pesanan aktif (harus urut ID krn inorder) ----
section("LIHAT PESANAN AKTIF (BST INORDER)")
sistem.lihat_pesanan_aktif()

# ---- 4. Cari pesanan ----
section("CARI PESANAN")
sistem.cari_pesanan(2)
sistem.cari_pesanan(99)  # tidak ada

# ---- 5. Dapur proses prioritas tertinggi ----
# Harapan: Take Away Express (Budi id2, Dedi id4) diproses duluan drpd Dine In,
# dan di antara sesama Take Away, yang lebih dulu masuk (Budi id2) duluan.
section("DAPUR PROSES PRIORITAS (HEAP)")
for _ in range(2):
    sistem.proses_prioritas_dapur()

section("STATISTIK SETELAH 2x PROSES DAPUR")
sistem.lihat_statistik()

# ---- 6. Batalkan pesanan ----
section("BATALKAN PESANAN (Citra, id3)")
sistem.batalkan_pesanan(3)
sistem.lihat_pesanan_aktif()

# ---- 7. Undo pembatalan terakhir ----
section("UNDO PEMBATALAN TERAKHIR")
sistem.undo_pembatalan_terakhir()
sistem.lihat_pesanan_aktif()

# ---- Coba undo lagi padahal top stack bukan pembatalan ----
section("COBA UNDO LAGI (harus gagal, top stack = Selesai)")
sistem.undo_pembatalan_terakhir()

# ---- 8. Lihat riwayat ----
section("RIWAYAT TRANSAKSI (STACK)")
sistem.lihat_riwayat()

# ---- 9. Proses sisa dapur & lihat statistik akhir ----
section("PROSES SISA DAPUR")
while not sistem.antrean_dapur.is_empty():
    sistem.proses_prioritas_dapur()

section("STATISTIK AKHIR")
sistem.lihat_statistik()

# ---- Skenario edge case ----
section("EDGE CASE: Antrean masuk kosong")
sistem.proses_antrean()

section("EDGE CASE: Antrean dapur kosong")
sistem.proses_prioritas_dapur()

section("EDGE CASE: Batalkan ID yang tidak ada")
sistem.batalkan_pesanan(999)

section("EDGE CASE: Riwayat kosong? (cek dulu)")
print("Riwayat size:", sistem.riwayat.size())

print("\n\nSEMUA TES SELESAI TANPA ERROR.")
````
