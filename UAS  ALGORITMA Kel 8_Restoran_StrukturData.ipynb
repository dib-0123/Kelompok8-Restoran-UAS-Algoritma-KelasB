### Ringkasan alur bisnis

1. **Queue** — pelanggan yang datang mengantre menunggu giliran dilayani (FIFO).
2. **Binary Search Tree (BST)** — begitu pesanan dibuat, data pesanan disimpan sebagai node aktif (key = ID pesanan) agar bisa dicari, ditambah, dan dihapus secara efisien.
3. **Binary Heap** — setiap pesanan yang masuk dapur diberi skor prioritas (pesanan Take Away Express mendapat bonus prioritas), sehingga dapur selalu memproses pesanan dengan prioritas tertinggi lebih dulu.
4. **Stack** — setiap pesanan yang selesai diproses dicatat sebagai riwayat transaksi (LIFO), dan bisa dibatalkan (undo) lewat operasi pop.

Traversal BST yang dipakai adalah **inorder**, karena BST di-key berdasarkan ID pesanan sehingga inorder traversal otomatis menghasilkan daftar pesanan aktif yang terurut rapi.

### 1. Queue — Antrean Pelanggan (implementasi linked list)

```python
# ---------------------------------------------------------
# 1. QUEUE (Antrean Pelanggan) - Implementasi Linked List
# ---------------------------------------------------------
class NodeQueue:
    def __init__(self, data):
        self.data = data
        self.next = None


class Queue:
    def __init__(self):
        self.front = None
        self.rear = None
        self._jumlah = 0

    def is_empty(self):
        return self.front is None

    def enqueue(self, data):
        node_baru = NodeQueue(data)
        if self.is_empty():
            self.front = node_baru
            self.rear = node_baru
        else:
            self.rear.next = node_baru
            self.rear = node_baru
        self._jumlah += 1

    def dequeue(self):
        if self.is_empty():
            return None
        node_depan = self.front
        self.front = self.front.next
        if self.front is None:
            self.rear = None
        self._jumlah -= 1
        return node_depan.data

    def peek(self):
        if self.is_empty():
            return None
        return self.front.data

    def jumlah(self):
        return self._jumlah

    def display(self):
        if self.is_empty():
            print("Antrean pelanggan kosong.")
            return
        temp = self.front
        no = 1
        print("--- Antrean Pelanggan ---")
        while temp is not None:
            d = temp.data
            print(f"{no}. {d['nama']} (rombongan {d['jumlah_orang']} orang)")
            temp = temp.next
            no += 1
```

### 2. Stack — Riwayat Transaksi (implementasi linked list)

```python
# ---------------------------------------------------------
# 2. STACK (Riwayat Transaksi) - Implementasi Linked List
# ---------------------------------------------------------
class NodeStack:
    def __init__(self, data):
        self.data = data
        self.next = None


class Stack:
    def __init__(self):
        self.top = None
        self._jumlah = 0

    def is_empty(self):
        return self.top is None

    def push(self, data):
        node_baru = NodeStack(data)
        node_baru.next = self.top
        self.top = node_baru
        self._jumlah += 1

    def pop(self):
        if self.is_empty():
            return None
        node_teratas = self.top
        self.top = self.top.next
        self._jumlah -= 1
        return node_teratas.data

    def peek(self):
        if self.is_empty():
            return None
        return self.top.data

    def jumlah(self):
        return self._jumlah

    def display(self):
        if self.is_empty():
            print("Riwayat transaksi kosong.")
            return
        temp = self.top
        no = 1
        print("--- Riwayat Transaksi (paling baru di atas) ---")
        while temp is not None:
            print(f"{no}. {temp.data}")
            temp = temp.next
            no += 1
```

### 3. Binary Search Tree — Data Pesanan Aktif

```python
# ---------------------------------------------------------
# 3. BINARY SEARCH TREE (Data Pesanan Aktif)
# ---------------------------------------------------------
class NodeBST:
    def __init__(self, pesanan):
        self.pesanan = pesanan
        self.kiri = None
        self.kanan = None


class BST:
    def __init__(self):
        self.root = None

    def insert(self, pesanan):
        self.root = self._insert(self.root, pesanan)

    def _insert(self, node, pesanan):
        if node is None:
            return NodeBST(pesanan)
        if pesanan.id_pesanan < node.pesanan.id_pesanan:
            node.kiri = self._insert(node.kiri, pesanan)
        elif pesanan.id_pesanan > node.pesanan.id_pesanan:
            node.kanan = self._insert(node.kanan, pesanan)
        return node

    def search(self, id_pesanan):
        return self._search(self.root, id_pesanan)

    def _search(self, node, id_pesanan):
        if node is None:
            return None
        if id_pesanan == node.pesanan.id_pesanan:
            return node.pesanan
        elif id_pesanan < node.pesanan.id_pesanan:
            return self._search(node.kiri, id_pesanan)
        else:
            return self._search(node.kanan, id_pesanan)

    def delete(self, id_pesanan):
        self.root, berhasil = self._delete(self.root, id_pesanan)
        return berhasil

    def _delete(self, node, id_pesanan):
        if node is None:
            return node, False
        if id_pesanan < node.pesanan.id_pesanan:
            node.kiri, berhasil = self._delete(node.kiri, id_pesanan)
        elif id_pesanan > node.pesanan.id_pesanan:
            node.kanan, berhasil = self._delete(node.kanan, id_pesanan)
        else:
            berhasil = True
            if node.kiri is None:
                return node.kanan, berhasil
            elif node.kanan is None:
                return node.kiri, berhasil
            else:
                pengganti = self._cari_min(node.kanan)
                node.pesanan = pengganti.pesanan
                node.kanan, _ = self._delete(node.kanan, pengganti.pesanan.id_pesanan)
        return node, berhasil

    def _cari_min(self, node):
        while node.kiri is not None:
            node = node.kiri
        return node

    def inorder(self):
        hasil = []
        self._inorder(self.root, hasil)
        return hasil

    def _inorder(self, node, hasil):
        if node is not None:
            self._inorder(node.kiri, hasil)
            hasil.append(node.pesanan)
            self._inorder(node.kanan, hasil)

    def tinggi(self):
        return self._tinggi(self.root)

    def _tinggi(self, node):
        if node is None:
            return -1
        return 1 + max(self._tinggi(node.kiri), self._tinggi(node.kanan))

    def hitung_node(self):
        return self._hitung_node(self.root)

    def _hitung_node(self, node):
        if node is None:
            return 0
        return 1 + self._hitung_node(node.kiri) + self._hitung_node(node.kanan)

    def display(self):
        daftar = self.inorder()
        if not daftar:
            print("Tidak ada pesanan aktif.")
            return
        print("--- Daftar Pesanan Aktif (terurut berdasarkan ID) ---")
        for p in daftar:
            print(p)
```

### 4. Binary Heap — Prioritas Dapur (implementasi array/list biasa, bukan `heapq`)

```python
# ---------------------------------------------------------
# 4. BINARY HEAP (Prioritas Dapur) - Implementasi Array (list biasa)
# ---------------------------------------------------------
class BinaryHeap:
    def __init__(self):
        self.heap = []  # list Python dipakai murni sebagai array, BUKAN heapq

    def is_empty(self):
        return len(self.heap) == 0

    def _parent(self, i):
        return (i - 1) // 2

    def _kiri(self, i):
        return 2 * i + 1

    def _kanan(self, i):
        return 2 * i + 2

    def insert(self, pesanan):
        self.heap.append(pesanan)
        self._heapify_up(len(self.heap) - 1)

    def _heapify_up(self, i):
        while i > 0:
            idx_induk = self._parent(i)
            if self.heap[i].priority_score < self.heap[idx_induk].priority_score:
                self.heap[i], self.heap[idx_induk] = self.heap[idx_induk], self.heap[i]
                i = idx_induk
            else:
                break

    def delete_root(self):
        if self.is_empty():
            return None
        akar = self.heap[0]
        elemen_terakhir = self.heap.pop()
        if not self.is_empty():
            self.heap[0] = elemen_terakhir
            self._heapify_down(0)
        return akar

    def _heapify_down(self, i):
        n = len(self.heap)
        while True:
            kiri = self._kiri(i)
            kanan = self._kanan(i)
            terkecil = i
            if kiri < n and self.heap[kiri].priority_score < self.heap[terkecil].priority_score:
                terkecil = kiri
            if kanan < n and self.heap[kanan].priority_score < self.heap[terkecil].priority_score:
                terkecil = kanan
            if terkecil == i:
                break
            self.heap[i], self.heap[terkecil] = self.heap[terkecil], self.heap[i]
            i = terkecil

    def peek(self):
        if self.is_empty():
            return None
        return self.heap[0]

    def display(self):
        if self.is_empty():
            print("Antrean prioritas dapur kosong.")
            return
        print("--- Antrean Prioritas Dapur (Binary Heap) ---")
        for i, p in enumerate(self.heap):
            print(f"{i}. {p}")
```

### 5. Kelas Pesanan

```python
# ---------------------------------------------------------
# 5. KELAS PESANAN
# ---------------------------------------------------------
class Pesanan:
    def __init__(self, id_pesanan, nama_pelanggan, no_meja, jenis, daftar_menu, waktu_masuk):
        self.id_pesanan = id_pesanan
        self.nama_pelanggan = nama_pelanggan
        self.no_meja = no_meja
        self.jenis = jenis
        self.daftar_menu = daftar_menu
        self.waktu_masuk = waktu_masuk
        self.status = "Menunggu diproses"
        self.total_harga = sum(harga for _, harga in daftar_menu)

        # Take Away Express dapat bonus prioritas (angka lebih kecil = diproses lebih dulu)
        bonus_prioritas = 1000 if jenis == "Take Away Express" else 0
        self.priority_score = waktu_masuk - bonus_prioritas

    def __str__(self):
        menu_str = ", ".join(nama for nama, _ in self.daftar_menu)
        return (f"ID {self.id_pesanan} | Meja {self.no_meja} | {self.nama_pelanggan} | "
                f"{self.jenis} | Menu: {menu_str} | Total: Rp{self.total_harga:,} | "
                f"Status: {self.status}")
```

### 6. Sistem Restoran — Integrasi Queue, BST, Heap, dan Stack

```python
# ---------------------------------------------------------
# 6. SISTEM RESTORAN (Integrasi Queue, BST, Heap, Stack)
# ---------------------------------------------------------
class SistemRestoran:
    def __init__(self):
        self.antrean_pelanggan = Queue()
        self.data_pesanan = BST()
        self.prioritas_dapur = BinaryHeap()
        self.riwayat_transaksi = Stack()
        self.counter_id = 1
        self.counter_waktu = 1
        self.menu_restoran = {
            "1": ("Nasi Goreng Spesial", 20000),
            "2": ("Ayam Bakar", 25000),
            "3": ("Mie Ayam", 15000),
            "4": ("Es Teh Manis", 5000),
            "5": ("Jus Alpukat", 12000),
        }

    def tampilkan_menu_makanan(self):
        print("--- Daftar Menu ---")
        for kode, (nama_menu, harga) in self.menu_restoran.items():
            print(f"{kode}. {nama_menu} - Rp{harga:,}")

    # Tahap 1: Pelanggan datang -> Queue
    def pelanggan_datang(self, nama, jumlah_orang):
        data = {"nama": nama, "jumlah_orang": jumlah_orang, "waktu": self.counter_waktu}
        self.antrean_pelanggan.enqueue(data)
        self.counter_waktu += 1
        print(f"{nama} berhasil masuk antrean (posisi ke-{self.antrean_pelanggan.jumlah()}).")

    # Tahap 2: Panggil antrean -> buat pesanan -> masuk BST & Heap
    def layani_pelanggan_berikutnya(self, no_meja, pilihan_jenis, kode_menu_list):
        if self.antrean_pelanggan.is_empty():
            print("Tidak ada pelanggan dalam antrean.")
            return None

        intip = self.antrean_pelanggan.peek()
        print(f"Antrean berikutnya: {intip['nama']}")
        data_pelanggan = self.antrean_pelanggan.dequeue()

        jenis = "Take Away Express" if pilihan_jenis == "2" else "Dine In"

        daftar_menu = []
        for kode in kode_menu_list:
            if kode in self.menu_restoran:
                daftar_menu.append(self.menu_restoran[kode])
            else:
                print(f"Kode menu '{kode}' tidak ditemukan, dilewati.")

        if not daftar_menu:
            print("Pesanan kosong, dibatalkan.")
            return None

        pesanan_baru = Pesanan(self.counter_id, data_pelanggan["nama"], no_meja,
                                jenis, daftar_menu, self.counter_waktu)
        self.counter_id += 1
        self.counter_waktu += 1

        self.data_pesanan.insert(pesanan_baru)
        self.prioritas_dapur.insert(pesanan_baru)

        print(f"Pesanan berhasil dibuat -> ID {pesanan_baru.id_pesanan}, masuk antrean dapur.")
        return pesanan_baru

    # Tahap 3: Dapur proses pesanan prioritas tertinggi -> Heap -> BST delete -> Stack
    def proses_dapur(self):
        if self.prioritas_dapur.is_empty():
            print("Tidak ada pesanan yang perlu diproses dapur.")
            return None

        intip = self.prioritas_dapur.peek()
        print(f"Pesanan prioritas tertinggi saat ini: {intip}")

        pesanan = self.prioritas_dapur.delete_root()
        pesanan.status = "Selesai dimasak & disajikan"

        self.data_pesanan.delete(pesanan.id_pesanan)
        self.riwayat_transaksi.push(pesanan)

        print(f"Dapur selesai memproses: {pesanan}")
        return pesanan

    # BST search
    def cari_pesanan(self, id_pesanan):
        hasil = self.data_pesanan.search(id_pesanan)
        if hasil:
            print(f"Pesanan ditemukan: {hasil}")
        else:
            print(f"Pesanan dengan ID {id_pesanan} tidak ditemukan (mungkin sudah selesai).")
        return hasil

    # Stack pop / undo
    def batalkan_transaksi_terakhir(self):
        if self.riwayat_transaksi.is_empty():
            print("Tidak ada riwayat transaksi untuk dibatalkan.")
            return None
        intip = self.riwayat_transaksi.peek()
        print(f"Transaksi yang akan dibatalkan: {intip}")
        pesanan = self.riwayat_transaksi.pop()
        print(f"Transaksi ID {pesanan.id_pesanan} berhasil dibatalkan.")
        return pesanan

    # Display
    def tampilkan_antrean(self):
        self.antrean_pelanggan.display()

    def tampilkan_pesanan_aktif(self):
        self.data_pesanan.display()
        print(f"Jumlah pesanan aktif : {self.data_pesanan.hitung_node()}")
        print(f"Tinggi BST saat ini  : {self.data_pesanan.tinggi()}")

    def tampilkan_prioritas_dapur(self):
        self.prioritas_dapur.display()

    def tampilkan_riwayat(self):
        self.riwayat_transaksi.display()
```

### 7. Tampilan menu & program utama (jalankan cell ini untuk memulai program)

```python
# ---------------------------------------------------------
# 7. TAMPILAN MENU & PROGRAM UTAMA (CLI)
# ---------------------------------------------------------
def tampilkan_menu_utama():
    print("\n===== SISTEM ANTREAN & PEMESANAN RESTORAN =====")
    print("1. Pelanggan baru datang (masuk antrean)")
    print("2. Layani pelanggan berikutnya (buat pesanan)")
    print("3. Dapur proses pesanan prioritas tertinggi")
    print("4. Cari pesanan aktif berdasarkan ID")
    print("5. Batalkan transaksi terakhir (undo)")
    print("6. Tampilkan antrean pelanggan")
    print("7. Tampilkan pesanan aktif (BST - inorder)")
    print("8. Tampilkan antrean prioritas dapur (Heap)")
    print("9. Tampilkan riwayat transaksi (Stack)")
    print("0. Keluar")


def main():
    sistem = SistemRestoran()

    while True:
        tampilkan_menu_utama()
        pilihan = input("Pilih menu: ").strip()

        if pilihan == "1":
            nama = input("Nama pelanggan: ").strip()
            jumlah_orang = int(input("Jumlah orang: ").strip())
            sistem.pelanggan_datang(nama, jumlah_orang)

        elif pilihan == "2":
            if sistem.antrean_pelanggan.is_empty():
                print("Tidak ada pelanggan dalam antrean.")
                continue
            no_meja = input("Nomor meja yang ditempati: ").strip()
            sistem.tampilkan_menu_makanan()
            print("Jenis pesanan: 1) Dine In   2) Take Away Express")
            pilihan_jenis = input("Pilih jenis pesanan: ").strip()
            print("Masukkan kode menu satu per satu, ketik 'selesai' jika sudah.")
            kode_menu_list = []
            while True:
                kode = input("Kode menu: ").strip()
                if kode.lower() == "selesai":
                    break
                kode_menu_list.append(kode)
            sistem.layani_pelanggan_berikutnya(no_meja, pilihan_jenis, kode_menu_list)

        elif pilihan == "3":
            sistem.proses_dapur()

        elif pilihan == "4":
            id_cari = int(input("Masukkan ID pesanan: ").strip())
            sistem.cari_pesanan(id_cari)

        elif pilihan == "5":
            sistem.batalkan_transaksi_terakhir()

        elif pilihan == "6":
            sistem.tampilkan_antrean()

        elif pilihan == "7":
            sistem.tampilkan_pesanan_aktif()

        elif pilihan == "8":
            sistem.tampilkan_prioritas_dapur()

        elif pilihan == "9":
            sistem.tampilkan_riwayat()

        elif pilihan == "0":
            print("Terima kasih, sistem ditutup.")
            break

        else:
            print("Pilihan tidak valid, silakan coba lagi.")


if __name__ == "__main__":
    main()
```
